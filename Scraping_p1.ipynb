```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import Select
import pandas as pd
```

```python
#Driver for Chrome
driver = webdriver.Chrome('D:\APRENDER\WebScraping\chromedriver.exe')  

#Url for scraping
driver.get("https://www.lobstersnowboards.com/shop")

#DF for data of scraping
lobster = ["Url", "Name", "Image", "Price", "Size"]
#Select only United States products
UnitedStates = Select(driver.find_element_by_id('country_list'))
#driver.find_elements_by_xpath("//select[@name='country_list']//option[@value='26']")

#Click on the option United States
UnitedStates.select_by_value('26')
#print [us.set_attribute("selected") for us in UnitedStates]
#UnitedStates.set_attribute("selected")

array_urls = []
array_names = []
array_images = []
array_prices=[]
array_sizes = []

# get titles products
titles_url = driver.find_elements_by_xpath("//div[@class='col-6 col-sm-3 product-grid']//a[1]")

# get urls
url = [y.get_attribute("href") for y in titles_url]

# array.append(valor), array.insert(i, valor), array.tolist()
for i in range(len(url)):
    driver.get(url[i])
    
    array_urls.append(url[i])
    
    names = driver.find_elements_by_xpath("//h1[@class='product-title']")
    name = [y.text for y in names]
    array_names.append(name[0])
    
    images = driver.find_elements_by_xpath("//div[@class='col-12 col-sm-6 d-md-none']//div[@class='main-view']//img[@class='img-responsive']")
    image =[g.get_attribute("src") for g in images]
    array_images.append(image[0])
    
    prices = driver.find_elements_by_xpath("//div[@class='product_price']")
    price =[p.text for p in prices]
    array_prices.append(price[0])
    #print("price: " )
    #print(price, '\n')
    
    sizes = driver.find_elements_by_xpath("//select[@id='size']//option")
    size =[s.text for s in sizes]
    temp_size = size[1:]
    array_sizes.append(temp_size)
    
    
print("=====================")
print("array_url: ", array_urls)
print("=====================")
print("array_names: ", array_names)
print("=====================")
print("array_images: ", array_images)
print("=====================")
print("array_prices: ", array_prices)
print("=====================")
print("array_sizes: ", array_sizes)

```

```python
#lobster = ["Url", "Name", "Image", "Price", "Size"]

data = {'Url':array_urls, 'Name': array_names, 'Image':array_images,'Price':array_prices, 'Size': array_sizes }
df_lobster = pd.DataFrame(data)
df_lobster
#print(data)
```
